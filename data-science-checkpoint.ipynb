### 📍 Location-Based Analysis

```python
# Count occurrences of each location
location_counts = df['Location'].value_counts().reset_index()
location_counts.columns = ['Location', 'Count']

# Set dark background
plt.style.use('dark_background')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x=location_counts['Location'], 
    y=location_counts['Count'], 
    palette=['lime'] * len(location_counts)
)

# Add text labels
for i, count in enumerate(location_counts['Count']):
    plt.text(i, count + 1, str(count), ha='center', color='white', fontsize=12)

# Customize plot appearance
plt.xlabel('Location', fontsize=14, color='white')
plt.ylabel('Number of Customers', fontsize=14, color='white')
plt.title('Customer Count by Location', fontsize=16, color='white')
plt.xticks(rotation=45, color='white')
plt.yticks(color='white')

# Show the plot
plt.show()
```

## 💳 Payment and Discounts

```python
# Count occurrences of each payment method
payment_counts = df['Payment Method'].value_counts()

# Set dark background
plt.style.use('dark_background')

# Create the pie chart
plt.figure(figsize=(8, 8))
colors = plt.cm.plasma(range(len(payment_counts)))  # Using Plasma colormap
plt.pie(
    payment_counts, 
    labels=payment_counts.index, 
    autopct='%1.1f%%', 
    colors=colors, 
    textprops={'color': 'white'},
    startangle=140
)

# Customize plot appearance
plt.title('Popular Payment Methods', fontsize=16, color='white')

# Show the plot
plt.show()
```

```python
# Group data by subscription status and calculate total purchase amount
subscription_data = df.groupby('Subscription Status')['Purchase Amount (USD)'].sum().reset_index()

# Set dark background
plt.style.use('dark_background')

# Create the figure
plt.figure(figsize=(10, 6))

# Create the bar plot
sns.barplot(
    x=subscription_data['Subscription Status'], 
    y=subscription_data['Purchase Amount (USD)'], 
    palette='viridis'
)

# Add text labels
for i, amount in enumerate(subscription_data['Purchase Amount (USD)']):
    plt.text(i, amount + 500, f'${amount:,.0f}', ha='center', color='white', fontsize=12)

# Customize plot appearance
plt.xlabel('Subscription Status', fontsize=14, color='white')
plt.ylabel('Total Purchase Amount (USD)', fontsize=14, color='white')
plt.title('Impact of Subscription on Purchases', fontsize=16, color='white')
plt.xticks(color='white')
plt.yticks(color='white')

# Show the plot
plt.show()
```

```python
# Set dark background
plt.style.use('dark_background')

# Create the figure
plt.figure(figsize=(10, 6))

# Create the scatter plot
scatter = plt.scatter(
    df['Age'], 
    df['Purchase Amount (USD)'], 
    c=df['Age'], 
    cmap='viridis', 
    edgecolors='white', 
    alpha=0.8
)

# Add color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Age', color='white')
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(cbar.ax.yaxis.get_ticklabels(), color='white')

# Customize plot appearance
plt.xlabel('Age', fontsize=14, color='white')
plt.ylabel('Purchase Amount (USD)', fontsize=14, color='white')
plt.title('Age vs. Spending Habits', fontsize=16, color='white')
plt.xticks(color='white')
plt.yticks(color='white')

# Show the plot
plt.show()
```
